```python
from sklearn.cluster import KMeans
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
pd= pd.read_csv('votes.csv', skipinitialspace = True)

pd
```

```python
#turn ? to null and drop null values
new_pd = pd.replace({'?':np.nan}).dropna()

#drop any row that has more than 2 null
#new_pd = pd.replace({'?':np.nan})
#new_pd = new_pd[new_pd.isnull().sum(axis=1) < 2]
new_pd
```

```python
new_pd = new_pd.replace({'democrat':1, 'republican': 2, 'n':3, 'y': 4})
new_pd
```

```python
parties = new_pd.iloc[:,0]
parties
```

```python
votes = new_pd.iloc[:,1:16]
votes
```

```python
#votes.fillna(0, inplace=True)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(votes)
```

```python
parties_predict = kmeans.labels_
parties_predict
```

```python
accuracy_score(parties, parties_predict)
```
