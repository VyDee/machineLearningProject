```python
import pandas as pd
import numpy as np
pd = pd.read_csv('votes.csv', skipinitialspace = True)
pd
```

```python
new_pd = pd.replace({'?': np.nan}).dropna()
new_pd
```

```python
new_pd = new_pd.replace({'democrat':0, 'republican': 1, 'n':0, 'y': 1})
new_pd
```

```python
#parties is first column
parties = new_pd.iloc[:,0]
parties
```

```python
#votes is the rest
votes = new_pd.iloc[:,1:16]
votes
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', max_iter=1000)
rfe = RFE(model, n_features_to_select=10) 
fit = rfe.fit(votes, parties)
```

```python
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
#votes = votes[["d","e"]] #2 features
#votes = votes[["d","e","l"]] #3 features
#votes = votes[["d","e","f","l"]] #4 features
#votes = votes[["d","e","f","l","o"]] #5 features
#votes = votes[["d","e","f","k","l","m","o"]]#7 features
#votes = votes[["d","e","f","j","k","l","m","o"]]#8 features
votes = votes[["d","e","f","g","j","k","l","m","o","p"]] #10 features
votes
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(votes)
```

```python
parties_predict = kmeans.labels_
parties_predict
```

```python
kmeans.n_iter_
```

```python
kmeans.inertia_ 
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(parties, parties_predict)
```
